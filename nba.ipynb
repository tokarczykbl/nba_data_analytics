```python
# Import Dependencies
import pandas as pd
import os
from unidecode import unidecode
from pathlib import Path
from pprint import pprint
# import matplotlib.pyplot as plt
# import numpy as np
# import hvplot.pandas
# from babel.numbers import format_currency
# from matplotlib.ticker import FuncFormatter
```

```python
# Importing Cities, States and Countries files for prep and merge
# Cities Import
file_name = Path("data_downloads/Countries_States_Cities/cities.csv")
cities_df = pd.read_csv(file_name)

# Renaming Cities columns
cities_df = cities_df.rename(columns={"id":"city_id",
                                      "name":"city_name",
                                      "state_id":"state_id",
                                      "state_code":"state_code",
                                      "country_id":"country_id",
                                      "country_code":"country_code",
                                      "latitude":"city_latitude",
                                      "longitude":"city_longitude"
                                      })

#States Import
file_name = Path("data_downloads/Countries_States_Cities/states.csv")
states_df = pd.read_csv(file_name)

# Renaming States Columns
states_df = states_df.rename(columns={"id":"state_id",
                                      "name":"state_name",
                                      "country_id":"country_id",
                                      "country_code":"country_code",
                                      "state_code":"state_code",
                                      "latitude":"state_latitude",
                                      "longitude":"state_longitude",
                                      })

# Creating a list of columns need for merge from states_df
states_columns_needed = ['state_id','state_name', 'state_latitude', 'state_longitude']

#Countries Import
file_name = Path("data_downloads/Countries_States_Cities/countries.csv")
countries_df = pd.read_csv(file_name)

# Renaming countries columns
countries_df = countries_df.rename(columns={"id":"country_id",
                                            "name":"country_name",
                                            "iso3":"country_iso3",
                                            "iso2":"country_iso2",
                                            "phone_code":"country_phone_code",
                                            "capital":"country_capital",
                                            "currency":"country_currency",
                                            "currency_symbol":"country_currency_symbol",
                                            "tld":"country_tld",
                                            "native":"country_native",
                                            "region":"country_region",
                                            "subregion":"country_subregion",
                                            "timezones":"country_timezones",
                                            "latitude":"country_latitude",
                                            "longitude":"country_longitude",
                                            "emoji":"country_emoji",
                                            "emojiU":"country_emojiU"
                                            })

# Merging files and creating a single master dataframe
cities_states_df = pd.merge(cities_df, states_df[states_columns_needed], on='state_id', how='left')

cities_states_countries_df =  pd.merge(cities_states_df, countries_df, how='left', on='country_id')


cities_states_countries_master_df = cities_states_countries_df[['city_name', 'city_latitude', 'city_longitude',
                                                                'state_name', 'state_code', 'state_latitude', 'state_longitude',
                                                                'country_name', 'country_code', 'country_latitude', 'country_longitude',
                                                                'country_iso3',
                                                                'country_iso2',
                                                                'country_phone_code',
                                                                'country_capital',
                                                                'country_currency',
                                                                'country_currency_symbol',
                                                                'country_tld',
                                                                'country_native',
                                                                'country_region',
                                                                'country_subregion',
                                                                'country_timezones',
                                                                'country_emoji',
                                                                'country_emojiU']]

# Sorting columns and reseting index
cities_states_countries_master_df = cities_states_countries_master_df.sort_values(by=['country_region', 'country_subregion', 'country_name', 'state_name', 'city_name']).reset_index(drop=True)

# creating locations dataframe for NBA analysis
cities_states_countries_nba_df = cities_states_countries_master_df[['city_name', 'city_latitude', 'city_longitude',
                                                                    'state_name','state_latitude', 'state_longitude',
                                                                    'country_name','country_latitude', 'country_longitude', 'country_iso3',
                                                                    'country_region', 'country_subregion',]].drop_duplicates()
```
